```python
import pandas as pd

# Charge nos donnÃ©es de foot
df = pd.read_csv('../data/raw/premier_league_matches.csv')

print(f"âœ… Loaded {len(df)} matches")
df.describe()
```

```python
match_8_buts = df[df['away_score'] == 8]
print(match_8_buts)
```

```python
# 1. Matchs oÃ¹ domicile a gagnÃ© avec 3+ buts
big_home_wins = df[df['home_score'] >= 3]
print(f"âœ… {len(big_home_wins)} matchs avec 3+ buts Ã  domicile\n")
print(big_home_wins[['home_team', 'away_team', 'home_score', 'away_score']].head())

print("\n" + "="*50 + "\n")

# 2. Matchs d'Arsenal Ã  domicile
arsenal_home = df[df['home_team'] == 'Arsenal FC']
print(f"âš½ Arsenal a jouÃ© {len(arsenal_home)} matchs Ã  domicile\n")
print(arsenal_home[['date', 'away_team', 'home_score', 'away_score']].head())

print("\n" + "="*50 + "\n")

# 3. Matchs nuls (draw)
draws = df[df['home_score'] == df['away_score']]
print(f"ðŸ¤ {len(draws)} matchs nuls\n")
print(draws[['home_team', 'away_team', 'home_score', 'away_score']].head())
```

```python
big_home_wins_only = df[(df['home_score'] >= 3) & (df['home_score'] > df['away_score'])]
print(f"ðŸ† {len(big_home_wins_only)} matchs avec 3+ buts ET victoire domicile")
print(big_home_wins_only[['home_team', 'away_team', 'home_score', 'away_score']].head())

print("\n" + "="*50 + "\n")

big_teams = df[(df['home_team'] == 'Manchester City FC') | (df['home_team'] == 'Liverpool FC')]
print(f"âš½ {len(big_teams)} matchs avec Man City ou Liverpool Ã  domicile")
print(big_teams[['date', 'home_team', 'away_team', 'home_score', 'away_score']].head())

print("\n" + "="*50 + "\n")

high_scoring = df[(df['home_score'] + df['away_score']) >= 6]
print(f"ðŸ”¥ {len(high_scoring)} matchs avec 6+ buts au total")
print(high_scoring[['home_team', 'away_team', 'home_score', 'away_score']])
```

```python
# 1. SÃ©lectionner UNE colonne
teams = df['home_team']
print(f"Type: {type(teams)}")  # Series
print(f"PremiÃ¨res Ã©quipes:\n{teams.head()}")

print("\n" + "="*50 + "\n")

# 2. SÃ©lectionner PLUSIEURS colonnes
match_summary = df[['home_team', 'away_team', 'home_score', 'away_score']]
print(f"Type: {type(match_summary)}")  # DataFrame
print(match_summary.head())

print("\n" + "="*50 + "\n")

# 3. Combiner sÃ©lection + filtrage
arsenal_results = df[df['home_team'] == 'Arsenal FC'][['date', 'away_team', 'home_score', 'away_score']]
print("Arsenal matchs:")
print(arsenal_results.head())
```

```python
# 1. Nombre de matchs par Ã©quipe Ã  domicile
matches_by_team = df.groupby('home_team').size()
print("ðŸ“Š Matchs Ã  domicile par Ã©quipe:")
print(matches_by_team.sort_values(ascending=False).head(10))

print("\n" + "="*50 + "\n")

# 2. Moyenne de buts par Ã©quipe Ã  domicile
avg_goals_home = df.groupby('home_team')['home_score'].mean()
print("âš½ Moyenne de buts Ã  domicile (Top 10):")
print(avg_goals_home.sort_values(ascending=False).head(10))

print("\n" + "="*50 + "\n")

# 3. Stats complÃ¨tes (count, mean, sum)
home_stats = df.groupby('home_team')['home_score'].agg(['count', 'mean', 'sum'])
home_stats.columns = ['Matches', 'Avg Goals', 'Total Goals']
print("ðŸ† Stats complÃ¨tes Ã  domicile (Top 10 scoreurs):")
print(home_stats.sort_values('Total Goals', ascending=False).head(10))
```

```python
# 1. Total de buts par match
df['total_goals'] = df['home_score'] + df['away_score']

# 2. Goal difference (pour l'Ã©quipe domicile)
df['goal_diff'] = df['home_score'] - df['away_score']

# 3. RÃ©sultat du match (H = Home win, A = Away win, D = Draw)
def get_result(row):
    if row['home_score'] > row['away_score']:
        return 'H'
    elif row['home_score'] < row['away_score']:
        return 'A'
    else:
        return 'D'

df['result'] = df.apply(get_result, axis=1)

# Voir les nouvelles colonnes
print("âœ… Nouvelles colonnes crÃ©Ã©es!")
print(df[['home_team', 'away_team', 'home_score', 'away_score', 'total_goals', 'goal_diff', 'result']].head(10))

print("\n" + "="*50 + "\n")

# Stats sur les rÃ©sultats
print("Distribution des rÃ©sultats:")
print(df['result'].value_counts())
```
